```python
# Import library

from transformers import AutoModel,AutoTokenizer
```

```python
# load tokenizer

tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")
```

```python
# load language Model

model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")
```

```python
# tokenize this sentence

tokens = tokenizer("Hello Everyone",return_tensors= "pt")
```

```python
# printing tokens

print(tokens)
```

```python
# process the token

output = model(**tokens)[0]
```

```python
# Print output

print(output)
```

```python
# checking the shape of tokenizer

output.shape
```

```python
# let's see how two words "Hello Everyone" is showing into 4 token as per the shape

for token in tokens['input_ids'][0]:
  print(tokenizer.decode(token))
```
